In this lab, you will build a simple budget app that tracks spending in different categories and can show the relative spending percentage on a graph.

```python
import math
class Category:
    def __init__(self,name):
        self.name=name
        self.ledger=[]
        self.total=0
    def deposit(self,amount, description=""):
        self.total+=amount
        self.ledger.append({'amount':amount, 'description':description })

    def withdraw(self,amount,description=""):
        if(self.check_funds(amount)):
            self.total-=amount
            self.ledger.append({'amount':-amount, 'description':description })
            return True
        else:
            return False

    def get_balance(self):
        return self.total

    def transfer(self, amount, category):
        if(self.check_funds(amount)):
            self.withdraw(amount,f'Transfer to {category.name}')
            category.deposit(amount,f'Transfer from {self.name}')
            return True
        else:
            return False
    def check_funds(self,amount):
        if(amount > self.total):
            return False
        else:
            return True

    def __str__(self):
        cat_len=len(self.name)
        #star_len=int((30-cat_len)/2)
        #category_str=(f'*'*star_len+self.name+'*'* star_len)
        category_str=self.name.center(30,"*")

        for line in self.ledger:
           desc=line['description']
           category_str+=f"\n{desc[0:23].ljust(23)}"+f"{line['amount']:.2f}".rjust(7)

        category_str+=f"\nTotal: {self.get_balance():.2f}"
        return category_str


        category_str+="\n Total: "+str(self.total)
        return category_str


def create_spend_chart(categories):
    chart="Percentage spent by category\n"
    cat_totals=[]
    all_total=0
    for c in categories:
        cat_total=0
        for l in c.ledger:
            if(l["amount"] < 0):
                cat_total+=l["amount"]
        all_total+=cat_total
        cat_totals.append({"name":c.name,"total_spent":cat_total})
    #print(cat_totals)
    #print(all_total)

    cat_withdraw_per=[]
    for t in cat_totals:
        cat_per= math.floor(10*t["total_spent"]/all_total)*10
        #round(((100*t["total_spent"])/all_total),-1)

        cat_withdraw_per.append({"name":t["name"],"per":cat_per})


    #print(cat_withdraw_per)

    for i in range(100,-1,-10):
        chart += str(i).rjust(3) + '|'
        for j in cat_withdraw_per:
            if(i<=j["per"]):
                chart+=' o '
            else:
                chart+='   '
        chart+=' \n'

    # Print the separation line
    #chart += '    ' + '-'*(3*len(cat_withdraw_per) + 1) + ' \n'
    chart +=' '*4+'-'*(3*len(cat_withdraw_per)+1)+'\n'

    max_name_length = max([len(cat["name"]) for cat in cat_withdraw_per])
    expanded_cat_names = [cat["name"].ljust(max_name_length) for cat in cat_withdraw_per]

    #print(expanded_cat_names)

    for i in range(max_name_length):
        chart += '    '
        for j in range(len(cat_withdraw_per)):
            chart += ' ' + expanded_cat_names[j][i] + ' '
        if i <= max_name_length - 1:
            chart += ' \n'
    chart=chart.rstrip('\n')
    return chart

food = Category('Food')
food.deposit(1000, 'initial deposit')
food.withdraw(10.15, 'groceries')
food.withdraw(15.89, 'restaurant and more food for dessert')
clothing = Category('Clothing')
food.transfer(50, clothing)
print(food)

print(create_spend_chart([food,clothing]))
```
