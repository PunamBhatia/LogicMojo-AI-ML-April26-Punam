```python
##Write a program to print a N x M matrix of *
try:
  rows=int(input("Enter number of rows in matrix"))
  cols= int(input("Enter number of cols in matrix"))
  print("Here is the ",rows," X ",cols, " matrix of *")
  for n in range(rows):
    for m in range(cols):
      print("*",end='')
    print()
except ValueError:
   print("Invalid input, please enter a number.")
```

```python
##stair pattern
for i in range(5):
  for j in range(i+1):
    print("*",end='')
  print()
```

```python
##reverse stair pattern
n=6
for i in range(n):
  for j in range(n-i):
    print("*",end='')
  print()
```

```python
##Vertical Pyramid Pattern
q= 3
r= 1 ##n%2

space=' '
for i in range(q):
  for sp in range(q-i):
    print(space, end='')

  for st in range (r+i):
    print("*",end='')
  r+=1

  for sp in range(q-i):
    print(space,end='')
  print()
```

```python
##Upside down pyramid pattern
q= 4
##i=q
r= 2*q-1 ##n%2

space=' '
for i in range(q):
  for sp in range(q+i):
    print(space, end='')

  for st in range (r):
    print("*",end='')
  r-=2

  for sp in range(q+i):
    print(space,end='')
  print()

```

```python
## diamon pattern

q= 5
r= 1 ##n%2

space='&'
for i in range(q):
  for sp in range(q-i):
    print(space, end='')

  for st in range (r+i):
    print("*",end='')
  r+=1

  for sp in range(q-i):
    print(space,end='')
  print()


d=q-1
r=2*d-1
for i in range(d):
  for sp in range(d+i):
    print(space, end='')

  for st in range (r):
    print("*",end='')
  r-=2

  for sp in range(d+i):
    print(space,end='')
  print()
```
