### Use Case: YouTube Ad Frequency Change

You're a data scientist at YouTube, you are tasked with evaluating the impact of a proposed change in the ad frequency during time slots. Currently, YouTube displays one ad every 15 minutes. The management is considering a change to show two ads within the same time frame. The key question is whether this change in ad frequency will have any significant impact on user behavior.

**Experimental Design:**

To conduct a hypothesis test, a well-designed experiment is crucial. Here's how the experiment is structured:

* Random Sampling:
   * Select a random sample of 20,000 individuals from the user base.
   * Divide them into two groups: a treatment group and a control group.
* Implementation:
   * The treatment group experiences the new system with two ads per time slot.
   * The control group continues with the old system, featuring one ad per time slot.

**Data Collection:**
* Before the Change: Gather data on user engagement metrics for a specified period before the ad frequency change. For example, collect data on average watch time, bounce rate, and user satisfaction ratings.
* After the Change: Gather the same metrics for an equivalent period after the ad frequency change.

**Hypotheses:**

To frame the hypothesis for this scenario:

* Null Hypothesis (H0):
    * The mean watch time in the treatment group is greater than or equal to the mean watch time in the control group.
    * H0: μ_treatment ≥ μ_control
* Alternate Hypothesis (Ha):
    * The mean watch time in the treatment group is less than the mean watch time in the control group.
    * Ha: μ_treatment < μ_control

**Statistical Test:**

A two-sample one-sided t-test is chosen for comparing the means of the two groups. This test is suitable for scenarios with relatively small sample sizes, which is common in experimental settings.

**Data Exploration:**

1. Identification of Groups:
   * Clearly define the treatment and control groups.
   * Ensure equal sizes in both groups to maintain a balanced setup for unbiased comparisons.
2. Outlier Removal Initiation:
   * Recognize the impact of extreme outliers on data reliability.
   * Initiate the removal of outliers, with a focus on extreme "watch time" values.
   * Enhance data reliability by addressing extreme values in the specified variable.

**Select the Appropriate Test:**

* Depending on the data's characteristics (e.g., normality, sample size), choose a statistical test. Common tests include:
* Paired t-test: If the data is normally distributed and the sample size is small.
* ANOVA: If comparing multiple groups (e.g., different ad frequencies).
* Chi-Square Test: For categorical data like user satisfaction ratings.

**Analysis:**

* After data exploration, you observe that the control group, on average, spends more time on YouTube than the treatment group. The mean watch time difference suggests a nearly half-hour disparity between the two groups.

**Hypothesis Testing:**

Now, apply the two-sample one-sided t-test to determine if this observed difference in mean watch times is statistically significant.

* Null Hypothesis (H0):

    * μ_treatment ≥ μ_control
* Alternate Hypothesis (Ha):
    * μ_treatment < μ_control

If the p-value from the t-test is less than a chosen significance level (e.g., 0.05), you may reject the null hypothesis in favor of the alternate hypothesis. This implies that there is sufficient evidence to conclude that the mean watch time in the treatment group is indeed less than the mean watch time in the control group.

**Interpretation:**

The experiment's outcome guides YouTube in making informed decisions about whether the proposed change in ad frequency will likely lead to a significant impact on user behavior, as indicated by changes in watch time.

### Introduction to Hypothesis Testing

* Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data. It provides a formal process to determine whether the evidence in the data is strong enough to reject a preconceived notion (the null hypothesis) in favor of an alternative hypothesis.

image.png



**Types of Hypotheses:**

* One-Tailed Hypothesis: Specifies a direction for the difference (e.g., H₁: μ > μ₀).
* Two-Tailed Hypothesis: Does not specify a direction for the difference (e.g., H₁: μ ≠ μ₀).

**Why it is used?**

* Hypothesis testing is a fundamental tool for making data-driven decisions, validating research findings, and managing uncertainty. It provides a structured approach to evaluating evidence, testing theories, and making informed choices across various domains.

#### **Key Concepts**

##### **Null Hypothesis (H₀)**

* The statement that there is no effect or no difference, and it is the assumption that the test seeks to challenge. It is usually a statement of equality (e.g., H0: μ = μ0). The null hypothesis serves as a starting point for hypothesis testing. It is a hypothesis of no effect or no change.



##### **Alternative Hypothesis (H₁ or Ha)**

* The statement that indicates the presence of an effect or a difference. It is what you aim to support through the data (e.g., H1: μ != μ0). It asserts that there is a significant difference, effect, or relationship in the population under study. The alternative hypothesis represents the claim or hypothesis that researchers are trying to support with evidence.

##### **Type I Error**



* A Type I error occurs when you reject the null hypothesis (H0) when it is actually true. In simpler terms, it's like saying something is different when it's not.

* Factors Affecting Type I Error:
    * Significance Level (α): The lower the significance level, the less likely you are to commit a Type I error. However, a lower significance level also increases the chance of a Type II error (failing to reject a false null hypothesis).
    * Sample Size: A larger sample size can reduce the likelihood of a Type I error.
    * Effect Size: A larger effect size (a bigger difference between the groups being compared) can also reduce the likelihood of a Type I error.

image.png

* Example:
    * Imagine you're testing a new drug to see if it improves test scores. The null hypothesis would be that the drug has no effect on test scores. If you conduct the experiment and find significant results, leading you to reject the null hypothesis, but the drug actually has no effect, you've committed a Type I error.       

##### **Type II Error**

* A Type II error occurs when you fail to reject the null hypothesis (H0) when it is actually false. In simpler terms, it's like saying there is no difference when there actually is.

* Factors Affecting Type II Error:
    * Significance Level (α): A lower significance level increases the chance of a Type II error.
    * Sample Size: A smaller sample size can increase the likelihood of a Type II error.
    * Effect Size: A smaller effect size (a smaller difference between the groups being compared) can also increase the likelihood of a Type II error.
    * Power of the Test: The power of a test is the probability of correctly rejecting a false null hypothesis. A higher power reduces the chance of a Type II error.

    image.png

* Example:
    * Continuing with the drug example, imagine the drug does improve test scores. However, your experiment fails to find a significant difference, leading you to retain the null hypothesis. In this case, you've committed a Type II error.




##### **P-value**

* The p-value is a probability that measures the evidence against the null hypothesis. It represents the likelihood of observing a result as extreme or more extreme than the one obtained, assuming the null hypothesis is true.

* How to Interpret the P-Value:
    * Small p-value (typically less than 0.05): This indicates strong evidence against the null hypothesis. It suggests that the observed result is unlikely to occur by chance if the null hypothesis were true.
    * Large p-value (typically greater than 0.05): This indicates weak evidence against the null hypothesis. It suggests that the observed result could be due to chance.

In general, a p-value below the significance level (α) is considered statistically significant, leading to the rejection of the null hypothesis.

* Key Points to Remember:
    * The p-value is not a measure of the size of the effect. A small p-value simply indicates that the observed effect is unlikely to be due to chance.
    * The p-value should not be interpreted in isolation. It should be considered in conjunction with the effect size and other factors.
    * A large p-value does not necessarily mean that the null hypothesis is true. It simply means that there is insufficient evidence to reject it.

    image.png

* Example:
    * Suppose you're testing a new fertilizer to see if it increases crop yield. The null hypothesis is that the fertilizer has no effect on yield. If you conduct an experiment and find a significant increase in yield with a p-value of 0.03, this means there's a 3% chance of observing such an increase by chance if the fertilizer were actually ineffective. Since the p-value is less than the typical significance level of 0.05, you would reject the null hypothesis and conclude that the fertilizer is likely effective.

##### **T-Table in Hypothesis Testing**

* In the context of hypothesis testing, a t-table is a statistical table that provides critical values for the t-distribution. It is used to determine the significance of a t-test, which is a statistical test used to compare means of two samples.

* Structure of a T-Table

   * A typical t-table is organized by degrees of freedom (df) and significance levels (α). The degrees of freedom are calculated as the total number of observations minus the number of groups being compared. The significance levels are typically 0.01, 0.05, and 0.10.
* How to Use a T-Table

   * Determine the degrees of freedom (df): Calculate the df based on your sample size and the number of groups being compared.
   * Choose the appropriate significance level: Select the α value that corresponds to the desired level of confidence.
   * Locate the critical value: Find the intersection of the df row and the α column in the t-table. This value is the critical value.
   * Compare the calculated t-statistic to the critical value: If the absolute value of your calculated t-statistic is greater than the critical value, you can reject the null hypothesis.

   ![picture](https://drive.google.com/uc?export=view&id=1Q1wBXcnWzCtrCeFqtctfkA3pn_flYMas)

* Limitations:
   * T-tables may have limitations for very high degrees of freedom, as the T distribution approaches the standard normal distribution (Z distribution).
   * Some T-tables may provide critical values only up to a certain degree of freedom, beyond which Z-table values are used.

* Example
   * Suppose you have two groups of 10 students each and you want to compare their average test scores. You calculate a t-statistic of 2.15. To determine if this is significant at a 0.05 level, you would look up the critical value in a t-table with 18 degrees of freedom (10 + 10 - 2). The critical value for a two-tailed test at α = 0.05 is 2.101. Since your calculated t-statistic (2.15) is greater than the critical value (2.101), you can reject the null hypothesis and conclude that there is a significant difference between the average test scores of the two groups.  

#### Importance of Confidence

* Type 1 error is deemed critical as it involves erroneously rejecting a true null hypothesis, potentially leading to adverse consequences such as punishing an innocent individual or making an incorrect decision.

* The significance level (α) represents the probability of committing a Type 1 error. The accurate relationship between α and the confidence level is expressed by: α = 1 - confidence level.

* For instance, a confidence level of 95% corresponds to α = 0.05. To mitigate the risk of Type 1 error, researchers or decision-makers often choose a high confidence level, such as 95% or 99%. However, it's crucial to note that this choice doesn't signify a percentage of confidence in the evidence supporting the null hypothesis; instead, it is associated with the precision of the interval estimate. While reducing the risk of Type 1 error, increasing confidence also introduces a potential trade-off, as a higher threshold for accepting evidence (lower α) may elevate the risk of Type 2 error, where a false null hypothesis is not rejected.

* Type II error occurs in hypothesis testing when you fail to reject a false null hypothesis. In simpler terms, it's when you miss a significant effect or difference that actually exists.

#### Power Analysis

* The power of a statistical test is the probability of correctly rejecting a false null hypothesis.

* Power = 1 - β, where β is the Type 2 error rate.

* Power indicates the test's ability to detect a true effect or difference when it exists in the population. A higher power implies a better chance of identifying a real effect.

* Power is influenced by the significance level (alpha), effect size, and sample size. A larger effect size and sample size contribute to increased power.

* Type 2 error can be controlled by increasing the sample size.

### Types of Hypothesis Testing

#### 1. Z-Test

* The Z-test is a statistical method used to determine if there is a significant difference between sample data and a known population parameter or between the means of two samples. It is typically used when the sample size is large and the population variance is known or the sample size is sufficiently large for the Central Limit Theorem to apply.

##### **Two-Sample Z-Test**

* The Two-Sample Z-Test is used to determine if there is a significant difference between the means of two independent samples. It is applicable when you have two separate groups and want to test if their means are significantly different from each other. This test assumes that the populations from which the samples are drawn have the same variance and that the sample sizes are sufficiently large for the sampling distribution of the sample means to be approximately normal.
* Formula:

image.png

Where:
   * x̄₁ and x̄₂: Sample means
   * μ₁ and μ₂: Population means
   * σ₁ and σ₂: Population standard deviations
   * n₁ and n₂: Sample sizes

Example
* A researcher wants to compare the average salaries of male and female employees in a company. A random sample of 100 male employees has a mean salary of 55,000, and a random sample of 100 female employees has a mean salary of 52,000. The population standard deviations for male and female salaries are known to be 10,000 and 8,000, respectively.

Hypotheses:

H₀: μ₁ = μ₂ (The average salaries of male and female employees are equal.)
H₁: μ₁ ≠ μ₂ (The average salaries of male and female employees are not equal.)
Significance Level: α = 0.05

Test Statistic:

Z = (55000 - 52000) - (0) / √(10000²/100 + 8000²/100) = 2.5

P-value:

Using a z-table, we find that the p-value for a two-tailed test with a z-statistic of 2.5 is approximately 0.0124.

Decision:

Since p-value (0.0124) ≤ α (0.05), we reject the null hypothesis. There is sufficient evidence to conclude that the average salaries of male and female employees are different.

**Visualizing a z-test in hypothesis testing can help you better understand the distribution of your test statistic and the significance of the results. A common way to visualize a z-test is by plotting the standard normal distribution (a bell curve), showing the z-score, and highlighting the critical regions or p-value areas.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for the z-test
mean_control = 0.30  # e.g., 30% conversion rate
mean_variation = 0.35  # e.g., 35% conversion rate
std_dev = 0.05  # Population standard deviation
n_control = 1000  # Sample size for control group
n_variation = 1200  # Sample size for variation group

# Calculate the z-score
z_score = (mean_variation - mean_control) / np.sqrt((std_dev**2 / n_control) + (std_dev**2 / n_variation))

# Generate a standard normal distribution
x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

# Plot the standard normal distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Shade the critical region (for a one-tailed test)
alpha = 0.05
z_critical = stats.norm.ppf(1 - alpha)  # z-critical value for one-tailed test at 95% confidence level

# Shade the critical region
plt.fill_between(x, 0, y, where=(x >= z_critical), color='red', alpha=0.3, label='Critical Region (α = 0.05)')

# Plot the z-score
plt.axvline(z_score, color='green', linestyle='--', lw=2, label=f'Z-Score = {z_score:.2f}')

# Annotate the z-score
plt.text(z_score + 0.1, max(y) * 0.8, f'Z-Score = {z_score:.2f}', color='green', fontsize=12)

# Add labels and title
plt.title('Z-Test Visualization in Hypothesis Testing')
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

* Explanation:
   * Standard Normal Distribution: The curve represents the distribution of z-scores under the null hypothesis, with a mean of 0 and a standard deviation of 1.

   * Z-Score Line: A green dashed line is plotted at the z-score calculated from your data. This line shows where your test statistic falls on the standard normal distribution.

   * Critical Region: The red shaded area on the right side of the distribution represents the critical region (for a one-tailed test). If your z-score falls in this region, the null hypothesis is rejected.

   * P-Value Area: If you were to shade the area under the curve starting from the z-score to the tail, that would represent the p-value. The smaller this area, the more significant your result.

#### 2. T-Test

* The T-test is a statistical method used to determine if there is a significant difference between the means of one or two groups. It is especially useful when sample sizes are small and the population variance is unknown. The T-test is more flexible than the Z-test in that it does not require the population variance to be known and can handle smaller sample sizes.

image.png

* Types of T-Tests:

##### **One-Sample T-Test**

* The One-Sample T-Test is a statistical test used to determine whether the mean of a single sample is significantly different from a known or hypothesized population mean. This test is especially useful when the sample size is small and the population variance is unknown.

* Purpose

   * To test if the mean of a sample is significantly different from a known population mean or a hypothesized value.

* Formula

   t = (x̄ - μ₀) / (s / √n)

where:

   * x̄: Sample mean
   * μ₀: Population mean
   * s: Sample standard deviation
   * n: Sample size

* Example

A company claims that their new light bulbs have a mean lifespan of 1000 hours. To test this claim, a sample of 25 bulbs is tested, and the mean lifespan is found to be 980 hours with a sample standard deviation of 100 hours.

Hypotheses:

H₀: μ = 1000
H₁: μ ≠ 1000
Significance Level: α = 0.05

Test Statistic:

t = (980 - 1000) / (100 / √25) = -1

P-value:

Using a t-table with 24 degrees of freedom, we find that the p-value for a two-tailed test with a t-statistic of -1 is approximately 0.33.

Decision:

Since p-value (0.33) > α (0.05), we fail to reject the null hypothesis. There is not enough evidence to conclude that the mean lifespan of the new light bulbs is different from 1000 hours.

##### **Independent Two-Sample T-Test**

* An independent two-sample t-test is used to compare the means of two independent samples. This test is appropriate when:

   * The samples are independent (e.g., data from two different groups).
   * Both samples have approximately normal distributions.
   * The population variances are assumed to be equal (homoscedasticity). If this assumption is not met, a Welch's t-test can be used.

* Steps in an Independent Two-Sample T-Test

   * State the Hypotheses:
      * Null Hypothesis (H₀): μ₁ = μ₂ (The means of the two populations are equal.)
      * Alternative Hypothesis (H₁): μ₁ ≠ μ₂ (The means of the two populations are not equal.)

  * Set the Significance Level (α):
  
   Choose a significance level (e.g., 0.05) to determine the critical region for rejecting the null hypothesis.
  * Calculate the Pooled Standard Deviation:

      * s_p = √((n₁ - 1)s₁² + (n₂ - 1)s₂²) / (n₁ + n₂ - 2)

      * s₁ and s₂: Sample standard deviations

      * n₁ and n₂: Sample sizes

  * Calculate the Test Statistic:

      * t = (x̄₁ - x̄₂) / (s_p * √(1/n₁ + 1/n₂))
      * x̄₁ and x̄₂: Sample means
  * Determine the P-value:

    Use a t-table or statistical software to find the p-value associated with the calculated t-statistic, considering the degrees of freedom (n₁ + n₂ - 2).

  * Make a Decision:

     * If p-value ≤ α, reject the null hypothesis.
     * If p-value > α, fail to reject the null hypothesis.


##### Paired Sample T-Test

* A paired sample t-test is used to compare the means of two dependent samples. This test is appropriate when:

   * The same individuals or objects are measured twice (e.g., before and after an intervention).
   * The differences between the paired measurements are normally distributed.
* Formula
  * t = d̄ / (s_d / √n)

where:
  * d̄: Mean of the differences
  * s_d: Standard deviation of the differences
  * n: Number of pairs

* Advantages:

   * Takes into account the pairing of data, which increases the test's sensitivity.
   * Useful for comparing measurements taken on the same subjects under different conditions.
* Disadvantages:

   * Assumes that differences are normally distributed, which may be problematic with very small samples.
   * The test only applies to paired or matched samples, not to independent groups.

##### **One-sided and Two-sided Hypotheses**

* In hypothesis testing, the terms "one-sided" and "two-sided" refer to the type of alternative hypothesis you are testing. Each type of hypothesis is used to test different kinds of questions about the data.

**One-Sided Hypotheses**

* Specifies a direction (greater than or less than) in the alternative hypothesis.

* Rejection criteria on one side, either greater than or less than.

**One Sample T-test Example:**

* Investigating if the average hemoglobin level in a sample of patients is greater than a known population mean of 12.8 g/dL.
   * Hypotheses
      * H0 : x <= 12.8 (Null hypothesis: No increase in mean)
      * H1 : x > 12.8 (Alternative hypothesis: Significant increase in mean)

**Two-Sample T-test Example:**

* Comparing the IQ scores of boys and girls, hypothesizing that the mean IQ score for girls is greater than the mean IQ score for boys.
   * Hypotheses
      * H0 : Mean IQ for boys >= Mean IQ for girls
      * H1 : Mean IQ for boys < Mean IQ for girls

**Two-sided Hypothesis**

* Alternative hypothesis states a difference without specifying the direction.

* Rejection criteria on both sides, looking for differences in either direction.

**One Sample T-test Example:**

* Testing if the average hemoglobin level in a sample of patients is different from a known population mean of 12.8 g/dL.
   * Hypotheses
      * H0 : x = 12.8 (Null hypothesis: No difference in means)
      * H1 : x ≠ 12.8 (Alternative hypothesis: Significant difference in means)
**Two-Sample T-test Example:**

* Comparing the IQ scores of boys and girls without specifying the direction of the expected difference.
   * Hypotheses
      * H0 : Mean IQ for boys = Mean IQ for girls
      * H1 : Mean IQ for boys ≠ Mean IQ for girls

**Visualizing a t-test in hypothesis testing involves plotting the t-distribution, marking the t-score, and highlighting the critical regions or p-value areas. This approach helps in understanding how the t-score compares to the critical value and whether the null hypothesis can be rejected.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters for the t-test
mean_control = 5.5  # Sample mean of control group
mean_variation = 6.0  # Sample mean of variation group
std_dev = 1.2  # Sample standard deviation (assumed to be the same for both groups)
n_control = 30  # Sample size for control group
n_variation = 30  # Sample size for variation group

# Degrees of freedom
df = n_control + n_variation - 2

# Calculate the t-score
pooled_std_dev = np.sqrt(((n_control - 1) * std_dev**2 + (n_variation - 1) * std_dev**2) / df)
t_score = (mean_variation - mean_control) / (pooled_std_dev * np.sqrt(1/n_control + 1/n_variation))

# Generate t-distribution
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, df)

# Plot the t-distribution
plt.figure(figsize=(10, 6))
plt.plot(x, y, label=f'T-Distribution (df={df})', color='blue')

# Determine the critical t-value for a two-tailed test at 95% confidence level
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

# Shade the critical regions for a two-tailed test
plt.fill_between(x, 0, y, where=(x <= -t_critical) | (x >= t_critical), color='red', alpha=0.3, label='Critical Region (α = 0.05)')

# Plot the t-score
plt.axvline(t_score, color='green', linestyle='--', lw=2, label=f'T-Score = {t_score:.2f}')

# Annotate the t-score
plt.text(t_score + 0.1, max(y) * 0.8, f'T-Score = {t_score:.2f}', color='green', fontsize=12)

# Add labels and title
plt.title('T-Test Visualization in Hypothesis Testing')
plt.xlabel('T-Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

* Explanation:
   * T-Distribution: The curve represents the t-distribution with df degrees of freedom, which accounts for the sample size.

   * T-Score Line: The green dashed line represents the calculated t-score, showing where your test statistic falls on the t-distribution.

   * Critical Region: The red shaded areas on both sides of the distribution represent the critical regions for a two-tailed test. If your t-score falls in these regions, you would reject the null hypothesis.

   * P-Value Area: The area under the curve beyond the t-score corresponds to the p-value. The smaller this area, the more significant the result.

#### 3. ANOVA (Analysis of Variance)

* ANOVA (Analysis of Variance) is a statistical technique used to determine whether there are significant differences between the means of three or more independent (or dependent) groups. It helps in comparing multiple groups to see if at least one group mean differs significantly from the others.

image.png

* Types of ANOVA

##### One-Way ANOVA

* Tests the effect of a single independent variable (factor) on a dependent variable.

**Steps:**

* State the Hypotheses:
   * Null Hypothesis (H₀): μ₁ = μ₂ = ... = μₖ (All group means are equal.)
   * Alternative Hypothesis (H₁): At least one group mean is different from the others.
* Set the Significance Level (α):
   * Choose a significance level (e.g., 0.05).
* Calculate the F-statistic:
   * F = (MS_between) / (MS_within)
      * MS_between: Mean square between groups
      * MS_within: Mean square within groups
* Determine the P-value:
   * Use an F-distribution table or statistical software to find the p-value associated with the calculated F-statistic, considering the degrees of freedom for between-group and within-group variations.
* Make a Decision:
   * If p-value ≤ α, reject the null hypothesis.
   * If p-value > α, fail to reject the null hypothesis.
* Post-Hoc Tests:
   * If the null hypothesis is rejected in a one-way ANOVA, post-hoc tests (e.g., Tukey's HSD, Bonferroni correction) can be used to determine which specific groups differ significantly from each other.

##### Two-Way ANOVA

* Tests the effect of two independent variables (factors) on a dependent variable, and also examines the interaction effect between the two factors.

**Steps:**

* State the Hypotheses:
   * Main Effect: Test whether there is a significant difference in the means of the levels of one factor, ignoring the other factor.
* Interaction Effect:
   * Test whether the effect of one factor depends on the levels of the other factor.
* Set the Significance Level (α):
   * Choose a significance level (e.g., 0.05).
* Calculate the F-statistics:
  * F for the first factor
  * F for the second factor
  * F for the interaction effect
* Determine the P-values:
   * Use an F-distribution table or statistical software to find the p-values associated with the calculated F-statistics.
* Make a Decision:
   * If p-value ≤ α, reject the corresponding null hypothesis.

##### Repeated Measures ANOVA

* Used when the same subjects are measured multiple times under different conditions. It accounts for the within-subject correlation.

**Steps:**

* State the Hypotheses:
   * Within-Subjects Effect: Test whether there is a significant difference in the means of the repeated measurements.
* Subject Effect:
   * Test whether there are individual differences among subjects.
* Set the Significance Level (α):
   * Choose a significance level (e.g., 0.05).
* Calculate the F-statistics:
   * F for the within-subjects effect
   * F for the subject effect
* Determine the P-values:
   * Use an F-distribution table or statistical software to find the p-values associated with the calculated F-statistics.
* Make a Decision:
   * If p-value ≤ α, reject the corresponding null hypothesis.

**Below is a Python example using matplotlib and seaborn to visualize ANOVA results with a boxplot, which is a common way to show group distributions and means.**

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Example data for three groups
np.random.seed(42)
group1 = np.random.normal(5, 1.5, 30)
group2 = np.random.normal(6, 1.5, 30)
group3 = np.random.normal(7, 1.5, 30)

# Perform ANOVA
f_statistic, p_value = stats.f_oneway(group1, group2, group3)

# Combine the data into a single DataFrame for visualization
import pandas as pd

data = pd.DataFrame({
    'Group 1': group1,
    'Group 2': group2,
    'Group 3': group3
})

# Convert to long-form DataFrame
data_melted = pd.melt(data, var_name='Group', value_name='Value')

# Plot the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Group', y='Value', data=data_melted, palette='Set2')

# Plot the overall mean
overall_mean = data_melted['Value'].mean()
plt.axhline(overall_mean, color='red', linestyle='--', label=f'Overall Mean = {overall_mean:.2f}')

# Add labels and title
plt.title('ANOVA Visualization: Group Means Comparison')
plt.ylabel('Value')
plt.xlabel('Group')

# Annotate F-statistic and p-value
plt.text(2.5, min(data_melted['Value']) - 0.5, f'F-Statistic = {f_statistic:.2f}\nP-Value = {p_value:.4f}',
         ha='center', color='black', fontsize=12)

plt.legend()
plt.grid(True)
plt.show()
```

**This example assumes that we have a dataset where we want to analyze the effect of two factors (e.g., "Drug Type" and "Dosage Level") on a response variable (e.g., "Blood Pressure Reduction").**

Let's create a synthetic dataset first:

```python
import pandas as pd
import numpy as np

# Create a synthetic dataset
np.random.seed(42)
data = {
    'Drug Type': np.repeat(['Drug A', 'Drug B', 'Drug C'], 30),
    'Dosage Level': np.tile(np.repeat(['Low', 'Medium', 'High'], 10), 3),
    'Blood Pressure Reduction': np.random.normal(loc=0, scale=1, size=90)
}

# Introduce an interaction effect
data['Blood Pressure Reduction'] += np.where((data['Drug Type'] == 'Drug A') & (data['Dosage Level'] == 'High'), 1.5, 0)
data['Blood Pressure Reduction'] += np.where((data['Drug Type'] == 'Drug B') & (data['Dosage Level'] == 'Medium'), 1.2, 0)
data['Blood Pressure Reduction'] += np.where((data['Drug Type'] == 'Drug C') & (data['Dosage Level'] == 'Low'), 1.0, 0)

df = pd.DataFrame(data)
```

You can use seaborn to create interaction plots, which help to visualize the interaction effects between the two factors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create an interaction plot using seaborn
plt.figure(figsize=(10, 6))
sns.pointplot(data=df, x='Dosage Level', y='Blood Pressure Reduction', hue='Drug Type',
              markers=["o", "s", "D"], linestyles=["-", "--", "-."])
plt.title('Interaction Plot for Two-Way ANOVA')
plt.xlabel('Dosage Level')
plt.ylabel('Mean Blood Pressure Reduction')
plt.legend(title='Drug Type')
plt.show()
```

### Hypothetical Scenarios

#### Comparison of Means

Evaluate differences in means between groups to compare their statistical significance.

Example: Investigating mean IQ scores of boys compared to girls.

  * Hypotheses
     * Null (H0): Mean IQ of boys is equal to the mean IQ of girls.
     * Alternate (Ha): Mean IQ of boys is not equal to the mean IQ of girls.
     
Test: T-test or Z-test.

#### Test of Proportions

Assess the impact of a change in proportions, particularly relevant in scenarios involving binary outcomes.

Example: Analyzing if a new website feature impacts conversion rates.

  * Hypotheses
     * Null (H0): The new feature has no impact on conversion rates.
     * Alternate (Ha): The new feature results in different conversion rates.
     
Test: Z-test for proportions.

#### Test of Independence

Examine the relationship between two categorical variables to determine if they are independent or associated.

Example: Investigating if coffee consumption is associated with the likelihood of developing cancer.

  * Hypotheses
      * Null (H0): Coffee consumption is independent of the occurrence of cancer.
      * Alternate (Ha): There is a dependency between coffee consumption and the likelihood of cancer.
      
Test: Chi-square test of Independence.

### Z Distribution and T Distribution

* The Z-distribution and T-distribution are two important probability distributions used in statistical hypothesis testing. Both are used to determine how much sample data deviates from the population mean, but they are applied in different contexts depending on the sample size and known characteristics of the population.

image.png

**Z Distribution**
* The Z-distribution, also known as the Standard Normal Distribution, is a special case of the normal distribution. It is used when the sample size is large or when the population variance is known.

* Characteristics:

   * Mean and Standard Deviation: The Z-distribution has a mean of 0 and a standard deviation of 1.
   * Shape: It is a symmetric, bell-shaped curve.
   * Use: Typically used when the sample size is large (n > 30) or when the population standard deviation is known.
* Z-Score Calculation:

   * The Z-score indicates how many standard deviations an element is from the mean. The formula for calculating the Z-score is:

 **𝑍=𝑋−𝜇/𝜎**


where:

  * X = Sample value
  * μ = Population mean
  * σ = Population standard deviation
* Applications:
  * Hypothesis Testing: Used for testing hypotheses about population means when the sample size is large or the population standard deviation is known.
  * Confidence Intervals: Construct confidence intervals for the population mean when the population standard deviation is known.
Example:
If the mean height of a population is 170 cm with a standard deviation of 10 cm, and you have a sample height of 180 cm, the Z-score would be:


**Z= 180−170/ 10 = 1.0**

This indicates the sample height is 1 standard deviation above the population mean.

**T-Distribution**

* The T-distribution is a family of distributions that are used when the sample size is small (n ≤ 30) or the population standard deviation is unknown. It accounts for the increased variability expected with smaller sample sizes.

* Characteristics:
   * Mean and Standard Deviation: The T-distribution has a mean of 0. Its standard deviation is larger than 1 and depends on the degrees of freedom.
   * Shape: It is bell-shaped and symmetric like the Z-distribution but has thicker tails, which accounts for the greater variability in smaller samples.
   * Degrees of Freedom (df): The shape of the T-distribution depends on the degrees of freedom, which are calculated as
df=n−1 for a single sample.
* T-Score Calculation:
   * The T-score is used to calculate how many standard deviations the sample mean is from the population mean, considering the sample size. The formula for the T-score is:

 **t = (x̄ - μ₀) / (s / √n)**

where
  * x̄: Sample mean
  * μ₀: Population mean
  * s: Sample standard deviation
  * n: Sample size

* Applications:

   * Hypothesis Testing: Used for testing hypotheses about population means when the sample size is small or the population standard deviation is unknown.
   * Confidence Intervals: Construct confidence intervals for the population mean when the sample standard deviation is used instead of the population standard deviation.
* Example:
If you have a sample mean of 100 cm with a sample standard deviation of 15 cm from a sample of 25 observations, and the population mean is hypothesized to be 105 cm, the T-score would be:

This indicates the sample mean is 1.67 standard errors below the hypothesized mean.

##### T Critical Values

* T critical values are specific thresholds or boundary values used in hypothesis testing with the T distribution.
* They indicate the point beyond which one would reject the null hypothesis in favor of the alternative hypothesis.
* The determination of T critical values is contingent upon two primary factors: the degrees of freedom (df) and the desired confidence level, often expressed as a percentage.
* Determining T Critical Values: T critical values are typically obtained from statistical tables known as T-tables. These tables provide a set of critical T-values corresponding to different degrees of freedom and confidence levels. Users locate the appropriate intersection of degrees of freedom and confidence level to find the critical T-value.
* Example:
   * Suppose we have an experiment with 20 degrees of freedom and a significance level of 0.05.
   * In the T-table, we locate the row with df = 20 and the column with alpha = 0.05. The corresponding cell provides the critical T-value.
   * If the calculated T statistic exceeds this critical value, we reject the null hypothesis.
* Decision Rule for Hypothesis Testing using T-Test:
    * If the Absolute Value of the Calculated T-Statistic is Greater Than or Equal to the T-Critical Value:
       * Decision: Reject the Null Hypothesis.
       * Interpretation: The observed data provides sufficient evidence to conclude that there is a statistically significant difference or effect, supporting the alternative hypothesis.
   * If the Absolute Value of the Calculated T-Statistic is Less Than the T-Critical Value:
       * Decision: Fail to Reject the Null Hypothesis.
       * Interpretation: The observed data does not provide enough evidence to reject the null hypothesis. The results are consistent with the possibility that any observed differences are due to random variability.

#### YouTube Ads Use Case

The goal is to assess the impact of a change in the number of ads shown per time slot on YouTube viewership. The change involves shifting from one ad per time slot (the old system) to two ads per time slot (the new system).

**Experimental Design:**

* Random Sampling: 20,000 individuals are randomly divided into two groups – a treatment group and a control group.
* Implementation: The treatment group is exposed to the new system (two ads per time slot), while the control group continues with the old system (one ad per time slot).
* Data Collection: Details such as customer ID, group assignment, date, and time spent on YouTube are collected daily.
* Hypotheses:
   * Null Hypothesis (H0): The mean watch time in the treatment group is greater than or equal to the mean watch time in the control group.
   * Alternate Hypothesis (H1): The mean watch time in the treatment group is less than the mean watch time in the control group.
   
Two-sample one-sided t-test. Chosen for comparing means of two groups with relatively small sample sizes.

Emphasis is placed on the t-test results, determining if the observed difference in mean watch times is statistically significant.

The experiment's conclusion is based on the statistical significance of the results. If the p-value is less than a chosen significance level (e.g., 0.05), the null hypothesis may be rejected in favor of the alternate hypothesis.
